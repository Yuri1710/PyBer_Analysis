## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type

total_rides_type = pyber_data_df.groupby(["type"]).count()["ride_id"],
total_rides_type
```

```python
# 2. Get the total drivers for each city type
drivers_by_city = city_data_df.groupby(["type"]).sum()["driver_count"]
drivers_by_city
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the DataFrame
pyber_data_df.head(20)
```

```python
#  3. Get the total amount of fares for each city type
sum_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
sum_fares_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride = sum_fares_by_type / total_rides_type
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = sum_fares_by_type / drivers_by_city
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 

summary = {"Total Rides": pd.Series(total_rides_type, index = ["Rural","Suburban", "Urban"]),
           "Total Drivers" : pd.Series(drivers_by_city, index = ["Rural","Suburban", "Urban"]),     
           "Total Fares" : pd.Series(sum_fares_by_type, index = ["Rural","Suburban", "Urban"]),
           "Average fare per ride" : pd.Series(average_fare_per_ride, index = ["Rural","Suburban", "Urban"]),        
           "Average fare per Driver" : pd.Series(average_fare_per_driver, index = ["Rural","Suburban", "Urban"])}


summary_df = pd.DataFrame(summary)
summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
summary_df.index.name = None

summary = {"Total Rides": pd.Series(total_rides_type, index = ["Rural","Suburban", "Urban"]),
           "Total Drivers" : pd.Series(drivers_by_city, index = ["Rural","Suburban", "Urban"]),     
           "Total Fares" : pd.Series(sum_fares_by_type, index = ["Rural","Suburban", "Urban"]),
           "Average fare per ride" : pd.Series(average_fare_per_ride, index = ["Rural","Suburban", "Urban"]),        
           "Average fare per Driver" : pd.Series(average_fare_per_driver, index = ["Rural","Suburban", "Urban"])}

summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_of_fares = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
sum_of_fares
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()

sum_of_fares = sum_of_fares.reset_index()
sum_of_fares
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

sum_of_fares_pivot = sum_of_fares.pivot(index = "date",columns = "type", values = "fare")

sum_of_fares_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

sum_of_fares_pivot.loc['2019-01-01' : '2019-04-30'].head(10)
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

sum_of_fares_pivot.index = pd.to_datetime(sum_of_fares_pivot.index)
sum_of_fares_pivot.info()
```

```python
# 7. Check that the datatype for the index is datetime using df.info()

sum_of_fares_pivot.index = pd.to_datetime(sum_of_fares_pivot.index)
sum_of_fares_pivot.info()
```

```python
sum_of_fares_pivot.columns
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
sum_of_fares_weekly = sum_of_fares_pivot.resample("W").sum()
sum_of_fares_weekly.head(10)
```

```python
sum_of_fares_pivot_Jan_Apr = sum_of_fares_pivot.loc["2019-01-01": "2019-04-30"]
sum_of_fares_pivot_Jan_Apr
```

```python
sum_of_fares_pivot_Jan_Apr.resample("W").sum().plot(kind="line", figsize=(15,5),lw=2.5)
plt.title("Total Fare by City Type", fontsize=20)
plt.savefig("analysis/PyBer_fare_summary.png")
plt.show
```
